```
  from tensorflow.keras.layers import Input, Add, Dense, Activation, ZeroPadding2D, BatchNormalization, Flatten, Conv2D, AveragePooling2D, MaxPooling2D, Dropout
  from tensorflow.keras.models import Model, load_model
  from tensorflow.keras import regularizers
  from tensorflow.keras.initializers import glorot_uniform
  from tensorflow.keras.utils import plot_model
```

```
def MyModel(rate,lbda, input_shape = (32, 32, 3), classes = 10, fc = [512]):

    X_input = Input(input_shape)

    X = Conv2D(filters = 32, kernel_size = (3, 3), strides = (1,1), padding = 'same', name = 'Conv1',kernel_regularizer=regularizers.l2(lbda), kernel_initializer = glorot_uniform)(X_input)
    X = BatchNormalization(axis = 3, name = 'Batch1')(X)   
    X = Activation('relu')(X)
    X = Dropout(rate)(X)
    
    X = Conv2D(filters = 32, kernel_size = (3, 3), strides = (1,1), padding = 'same', name = 'Conv2',kernel_regularizer=regularizers.l2(lbda), kernel_initializer = glorot_uniform)(X_input)
    X = BatchNormalization(axis = 3, name = 'Batch2')(X)   
    X = Activation('relu')(X)
    X = Dropout(rate)(X)
    
    X = MaxPooling2D((2, 2), strides=(2, 2))(X)
    
    
    X = Conv2D(filters = 64, kernel_size = (3, 3), strides = (1,1), padding = 'same', name = 'Conv3',kernel_regularizer=regularizers.l2(lbda), kernel_initializer = glorot_uniform)(X)
    X = BatchNormalization(axis = 3, name = 'Batch3')(X)   
    X = Activation('relu')(X)
    X = Dropout(rate)(X)
    
    X = Conv2D(filters = 64, kernel_size = (3, 3), strides = (1,1), padding = 'same', name = 'Conv4',kernel_regularizer=regularizers.l2(lbda), kernel_initializer = glorot_uniform)(X)
    X = BatchNormalization(axis = 3, name = 'Batch4')(X)   
    X = Activation('relu')(X)
    X = Dropout(rate)(X)
    
    X = MaxPooling2D((2, 2), strides=(2, 2))(X)
    
    
    X = Flatten()(X)

    for i,F in enumerate(fc):
      X = Dense (F, activation = 'relu', name = 'FC' + str(i+1),kernel_regularizer=regularizers.l2(lbda), kernel_initializer = glorot_uniform)(X)
      X = Dropout(rate)(X)
    X = Dense(classes, activation='softmax', name='Output',kernel_regularizer=regularizers.l2(lbda), kernel_initializer = glorot_uniform)(X)

    model = Model(inputs = X_input, outputs = X, name='ResNet50')

    return model
```

```
model = MyModel(rate = 0.4,lbda = 0.05, fc = [512])
Adam = tf.keras.optimizers.Adam(learning_rate=0.001, beta_1=0.9, beta_2=0.999, amsgrad=False)
model.compile(optimizer=Adam, loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```
def train_augmented (model):
  with tf.device('/device:GPU:0'):
    history = model.fit(X_train_augmented, Y_train_augmented, validation_data=(X_dev_original, Y_dev_original), epochs = 100, batch_size = 32,verbose = 0)

    plt.plot(history.history['acc'])
    plt.plot(history.history['val_acc'])
    plt.title('Model accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Test'], loc='upper left')
    plt.show()

    # Plot training & validation loss values
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('Model loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Test'], loc='upper left')
    plt.show()

def train_not_augmented (model):
  with tf.device('/device:GPU:0'):
    history = model.fit(X_train, Y_train, validation_data=(X_dev_original, Y_dev_original), epochs = 100, batch_size = 32,verbose = 0)

    plt.plot(history.history['acc'])
    plt.plot(history.history['val_acc'])
    plt.title('Model accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Test'], loc='upper left')
    plt.show()

    # Plot training & validation loss values
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('Model loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Test'], loc='upper left')
    plt.show()
```

```
def loop_trainings(lbda_axis,rate_axis,fc_axis):
  for lbda_val in lbda_axis:
    for rate_val in rate_axis:
      for fc_val in fc_axis:            
        model = MyModel(rate = rate_val,lbda = lbda_val, fc = fc_val)
        Adam = tf.keras.optimizers.Adam(learning_rate=0.001, beta_1=0.9, beta_2=0.999, amsgrad=False)
        model.compile(optimizer=Adam, loss='categorical_crossentropy', metrics=['accuracy'])
        #model.summary()
        train_not_augmented(model)
        print(f"Dataset: Not augmented")
        print(f"Lambda: {lbda_val}")
        print(f"Dropout rate: {rate_val}")
        print(f"FC layers: {fc_val}")
        train_augmented(model)
        print(f"Dataset: Augmented")
        print(f"Lambda: {lbda_val}")
        print(f"Dropout rate: {rate_val}")
        print(f"FC layers: {fc_val}")
```

```
loop_trainings([0,0.1],[0,0.4],[[512],[1024,512]])
```
